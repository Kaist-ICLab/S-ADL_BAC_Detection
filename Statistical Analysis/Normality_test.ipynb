```python
#정규성 돌린 것을 따로 뽑을때
Normality_table = pd.DataFrame(columns=normality_column_names, index= normality_row_names)

왜도 = df[df.iloc[:,1] == "왜도(W)"].iloc[:,3]
첨도 = df[df.iloc[:,1] == "첨도(K)"].iloc[:,3]
정규성 = df[df.iloc[:,0] == "정규성 검정"]

new_index=[]
for i in 정규성.index:
    new_index.append(i+3)
    new_index.append(i+4)
    new_index.append(i+5)

df_table = pd.DataFrame(index= new_index)
kolmo_stat_table = pd.DataFrame(index= new_index)
kolmo_prob_table = pd.DataFrame(index= new_index)
shapiro_stat_table = pd.DataFrame(index= new_index)
shapiro_prob_table = pd.DataFrame(index= new_index)

for i in new_index:
    df_table.loc[i,0] = df.iloc[i,2]
    kolmo_stat_table.loc[i,0] = df.iloc[i,1]
    kolmo_prob_table.loc[i,0] = df.iloc[i,3]
    shapiro_stat_table.loc[i,0] = df.iloc[i,4]
    shapiro_prob_table.loc[i,0] = df.iloc[i,6]

왜도 = 왜도.reset_index()
첨도 = 첨도.reset_index()
df_table = df_table.reset_index()
kolmo_stat_table = kolmo_stat_table.reset_index()
kolmo_prob_table = kolmo_prob_table.reset_index()
shapiro_stat_table = shapiro_stat_table.reset_index()
shapiro_prob_table = shapiro_prob_table.reset_index()

concat_df= pd.concat([왜도, 첨도, df_table, kolmo_stat_table,kolmo_prob_table,shapiro_stat_table,shapiro_prob_table], axis=1, ignore_index=True)
concat_df = concat_df.drop([0,2,4,6,8,10,12], axis=1)
concat_df.index = normality_row_names
concat_df.columns = normality_column_names
print(concat_df)

#데이터에 따라 sum 대신 다른 이름을 넣으면 됨.
concat_df.to_excel(result_addr+"sum"+filename)
    
```
