```python
import pandas as pd
import numpy as np
```

```python
group0=["total_log",
"Input_Stream__(C+IF+INF+F)",
"Fix__(F)",
"Incorrect_Fixed__(IF)",
"Incorrect_Not_Fixed__(INF)",
"All_Actions__(shift, 한영, 특수문자)",
"Input_Stream_with_all_actions__(ISø)",
"total_time__(Seconds)",
"Transcripted_text__(T)",]
group1= ["Word_per_second__(WPS)",
"Word_per_minute__(WPM)",
"AdjWPS",
"AdjWPM",
"Character_per_second__(CPS)",
"Keystrokes_per_second__(KSPS)",
"Gesture_per_second_(GPS)",]
group2=["average_intercharacter_time",
"median_intercharacter_time",
"min_intercharacter_time",
"max_intercharacter_time",]
group3=["Keystrokes_per_character__(KSPC)",
"Gesture_per_character__(GPC)",
"Minimum_string_distance__(MSD)",
"Corrected_error_rates",
"Uncorrected_error_rates",
"Total_error_rates",]
group4=["Utilized_bandwidth",
"Wasted_bandwidth",
"Cost_per_correction",]
```

```python
filename = "Multivariate_04_ANOVA.xlsx"
file_addr = "./HCI_Data/ANOVA_SPSS_mean/"
result_addr = "./HCI_Data/result/"
# summary_start_row= 2
# # summary_start_column= 2
# row_offset = 140
# variable_offset = 13

num_var = len(group0) +len(group4)
df = pd.read_excel(file_addr+filename, engine='openpyxl')
```

```python

row_names= [group0+group4]
column_names =["Mauchly’s Test of Sphericity (P>0.05)","Test of Within-Subjects Effects (P<0.05)","Multivariate Tests (P<0.05)","Pairwise Comparisons(0<->0.03)(P<0.05)","Pairwise Comparisons(0<->0.08)(P<0.05)","Pairwise Comparisons(0.03<->0.08)(P<0.05)","Within-Subjects Contrasts(Level 1 vs Level 2)","Within-Subjects Contrasts(Level 2 vs Level 3)"]
```

```python
Uni_ANOVA_table = pd.DataFrame(index= row_names)


#Mauchly
con_Mauchly = df[df.iloc[:,3] == "근사 카이제곱검정"]
Mauchly_table = df.iloc[con_Mauchly.index[0]+2:con_Mauchly.index[0]+2+num_var,5].reset_index().drop("index", axis=1)

#With-in
con_within = df[df.iloc[:,0] == "일변량 검정"]
within_table = pd.DataFrame()
for i in range(num_var):
    if (Mauchly_table.iloc[i,0]>=0.05):
        within_table.loc[i,0] = df.iloc[con_within.index[0] + 2+ 4*i,7]
    else:
        within_table.loc[i,0] = df.iloc[con_within.index[0] + 2+ 4*i+1,7]
within_table = within_table.reset_index().drop("index", axis=1)

#Multi variate
multi_1 = df[df.iloc[:,0] == "다변량 검정a"]
multi_2 = df[df.iloc[:,0] == "다변량a,b"]
multi = min(df.iloc[multi_1.index[0]+6,7], df.iloc[multi_2.index[0]+2,6])
multi_table = within_table.copy()
multi_table[0] = multi

#pairwise
con_pair = df[df.iloc[:,0] == "대응별 비교"]
pair_idx = con_pair.index[0] +3
pair_table = pd.DataFrame()
for i in range(num_var):
    pair_table.loc[i,0] = df.iloc[pair_idx + 6*i,5]
    pair_table.loc[i,1] = df.iloc[pair_idx + 6*i+1,5]
    pair_table.loc[i,2] = df.iloc[pair_idx + 6*i+3,5]
pair_table = pair_table.reset_index().drop("index", axis=1)

con_within_contrast = df[df.iloc[:,0] == "개체-내 대비 검정"]
within_contrast_index=con_within_contrast.index[0]+2

within_contrast_table = pd.DataFrame()
for i in range(num_var):
    within_contrast_table.loc[i,0] = df.iloc[within_contrast_index +2*i,7]
    within_contrast_table.loc[i,1] = df.iloc[within_contrast_index +2*i+1,7]
within_contrast_table = within_contrast_table.reset_index().drop("index", axis=1)
# print(within_contrast_table)
```

```python
concat_df= pd.concat([Mauchly_table,within_table,multi_table ,pair_table,within_contrast_table], axis=1, ignore_index=True)
concat_df.index = row_names
concat_df.columns = column_names
print(concat_df)
concat_df.to_excel(result_addr+"new_mean"+filename)
    
```
