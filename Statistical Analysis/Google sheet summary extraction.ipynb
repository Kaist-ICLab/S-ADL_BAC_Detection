```python
import pandas as pd
import numpy as np
```

```python
#불러올 파일 이름
filename = "NKEY_LOG(weather_searching)_E그룹.xlsx"
#파일 위치
file_addr = "./HCI_Data/"
#저장할 위치
result_addr = "./HCI_Data/result"
#구글 시트에서 summary 시작 행, 열
summary_start_row= 199
summary_start_column= 1
#구글 시트에서 trial 사이 offset
summary_trial_offset = 4
#추출할 변수 개수
num_var = 33

df = pd.read_excel(file_addr+filename, engine='openpyxl', sheet_name=None)
#시트 이름
sheet_names = list(df.keys())[2:]
print(sheet_names)
#anova 열 이름
ANOVA_column_names =['average_intercharacter_time_0', 'average_intercharacter_time_0.03', 'average_intercharacter_time_0.09', 'std_intercharacter_time_0', 'std_intercharacter_time_0.03', 'std_intercharacter_time_0.09', 'median_intercharacter_time_0', 'median_intercharacter_time_0.03', 'median_intercharacter_time_0.09', 'min_intercharacter_time_0', 'min_intercharacter_time_0.03', 'min_intercharacter_time_0.09', 'max_intercharacter_time_0', 'max_intercharacter_time_0.03', 'max_intercharacter_time_0.09', 'total_time(S)_0', 'total_time(S)_0.03', 'total_time(S)_0.09', 'total_count_0', 'total_count_0.03', 'total_count_0.09', 'IS_(C+IF+INF+F)_0', 'IS_(C+IF+INF+F)_0.03', 'IS_(C+IF+INF+F)_0.09', 'backspace_(F)_0', 'backspace_(F)_0.03', 'backspace_(F)_0.09', 'C_0', 'C_0.03', 'C_0.09', 'IF_0', 'IF_0.03', 'IF_0.09', 'INF_0', 'INF_0.03', 'INF_0.09', 'Shift_0', 'Shift_0.03', 'Shift_0.09', 'ISø_(IS+Shift)_0', 'ISø_(IS+Shift)_0.03', 'ISø_(IS+Shift)_0.09', 'T_(total_count-F)_0', 'T_(total_count-F)_0.03', 'T_(total_count-F)_0.09', 'WPS_0', 'WPS_0.03', 'WPS_0.09', 'WPM_0', 'WPM_0.03', 'WPM_0.09', 'AdjWPS_0', 'AdjWPS_0.03', 'AdjWPS_0.09', 'AdjWPM_0', 'AdjWPM_0.03', 'AdjWPM_0.09', 'CPS_0', 'CPS_0.03', 'CPS_0.09', 'KSPS_0', 'KSPS_0.03', 'KSPS_0.09', 'GPS_0', 'GPS_0.03', 'GPS_0.09', 'KSPC_0', 'KSPC_0.03', 'KSPC_0.09', 'GPC_0', 'GPC_0.03', 'GPC_0.09', 'MSD_0', 'MSD_0.03', 'MSD_0.09', 'Corrected_error_rates_0', 'Corrected_error_rates_0.03', 'Corrected_error_rates_0.09', 'Uncorrected_error_rates_0', 'Uncorrected_error_rates_0.03', 'Uncorrected_error_rates_0.09', 'Total_error_rates_0', 'Total_error_rates_0.03', 'Total_error_rates_0.09', 'Correction_efficiency_0', 'Correction_efficiency_0.03', 'Correction_efficiency_0.09', 'Participant_conscientiousness_0', 'Participant_conscientiousness_0.03', 'Participant_conscientiousness_0.09', 'Utilized_bandwidth_0', 'Utilized_bandwidth_0.03', 'Utilized_bandwidth_0.09', 'Wasted_bandwidth_0', 'Wasted_bandwidth_0.03', 'Wasted_bandwidth_0.09', 'Cost_per_correction_0', 'Cost_per_correction_0.03', 'Cost_per_correction_0.09']
```

```python
original_column_name = df['P013'][summary_start_row:].iloc[:,1]
# 전체(entire) 데이터 뽑을 때 열 이름
column_names =[]
for i in original_column_name:
    column_names.append(i+"_trial1")
    column_names.append(i+"_trial2")
    column_names.append(i+"_trial3")
    column_names.append(i+"_trial4")
    column_names.append(i+"_trial5")
    column_names.append(i+"_trial6")
    column_names.append(i+"_trial7")
    column_names.append(i+"_trial8")
    column_names.append(i+"_trial9")
```

```python
#mean~ cv 뽑는 코드

ANOVA_mean_table= pd.DataFrame(columns=ANOVA_column_names, index= sheet_names)
ANOVA_median_table= pd.DataFrame(columns=ANOVA_column_names, index= sheet_names)
ANOVA_min_table= pd.DataFrame(columns=ANOVA_column_names, index= sheet_names)
ANOVA_max_table= pd.DataFrame(columns=ANOVA_column_names, index= sheet_names)
ANOVA_sum_table= pd.DataFrame(columns=ANOVA_column_names, index= sheet_names)
ANOVA_std_table= pd.DataFrame(columns=ANOVA_column_names, index= sheet_names)
ANOVA_cv_table= pd.DataFrame(columns=ANOVA_column_names, index= sheet_names)

#시트 당 for문
for i in sheet_names:


    ############## 사이트 주소 - summary_start_row: summary_start_row+num_var#####################
    ############## 지역이름 - summary_start_row + num_var +1: summary_start_row + 2*num+var +1 #####################
    ############## 온도습도 - summary_start_row + 2*num_var +2: summary_start_row + 3*num+var +2 #####################
    summary = df[i][summary_start_row:summary_start_row + num_var ]
    
    
    ##DIV/0!을 NA로 변환
    summary = summary.replace("#DIV/0!", pd.NA)
    #첫 5개(average_character_time ~ )을 milsec에서 sec으로 바꿔줌
    summary.iloc[:5,range(summary_start_column, summary.shape[1], summary_trial_offset)] /= 1000 
    
    concat_summary = pd.DataFrame()
    #trial 별 summary를 1개의 연속된 dataframe으로 만듦.
    for j in range(summary_start_column, summary.shape[1], summary_trial_offset):
        concat_summary = pd.concat([concat_summary, summary.iloc[:,j]], axis=1)
    print(concat_summary)
    
    #mean
    calculated_summary = pd.DataFrame()
    calculated_summary["Trial 123"] = concat_summary.iloc[:num_var,:3].mean(axis=1)
    calculated_summary["Trial 456"] = concat_summary.iloc[:num_var,3:6].mean(axis=1)
    calculated_summary["Trial 789"] = concat_summary.iloc[:num_var,6:].mean(axis=1)
    ANOVA_mean_table.loc[i] = np.array(calculated_summary).reshape(1,-1)
    
    #median
    calculated_summary = pd.DataFrame()
    calculated_summary["Trial 123"] = concat_summary.iloc[:num_var,:3].median(axis=1)
    calculated_summary["Trial 456"] = concat_summary.iloc[:num_var,3:6].median(axis=1)
    calculated_summary["Trial 789"] = concat_summary.iloc[:num_var,6:].median(axis=1)
    ANOVA_median_table.loc[i] = np.array(calculated_summary).reshape(1,-1)
    
    #min
    calculated_summary = pd.DataFrame()
    calculated_summary["Trial 123"] = concat_summary.iloc[:num_var,:3].min(axis=1)
    calculated_summary["Trial 456"] = concat_summary.iloc[:num_var,3:6].min(axis=1)
    calculated_summary["Trial 789"] = concat_summary.iloc[:num_var,6:].min(axis=1)
    ANOVA_min_table.loc[i] = np.array(calculated_summary).reshape(1,-1)
    
    #max
    calculated_summary = pd.DataFrame()
    calculated_summary["Trial 123"] = concat_summary.iloc[:num_var,:3].max(axis=1)
    calculated_summary["Trial 456"] = concat_summary.iloc[:num_var,3:6].max(axis=1)
    calculated_summary["Trial 789"] = concat_summary.iloc[:num_var,6:].max(axis=1)
    ANOVA_max_table.loc[i] = np.array(calculated_summary).reshape(1,-1)
    
    #std
    calculated_summary = pd.DataFrame()
    calculated_summary["Trial 123"] = concat_summary.iloc[:num_var,:3].std(axis=1)
    calculated_summary["Trial 456"] = concat_summary.iloc[:num_var,3:6].std(axis=1)
    calculated_summary["Trial 789"] = concat_summary.iloc[:num_var,6:].std(axis=1)
    ANOVA_std_table.loc[i] = np.array(calculated_summary).reshape(1,-1)
   
    #cv
    
    normal = ANOVA_mean_table.loc[i] != 0
    abnormal = ANOVA_mean_table.loc[i] == 0
    ANOVA_cv_table.loc[i] = ANOVA_std_table.loc[i].copy()
    ANOVA_cv_table.loc[i][normal] /= ANOVA_mean_table.loc[i][normal]
    ANOVA_cv_table.loc[i][abnormal] = pd.NA    
    
    #sum
    calculated_summary = pd.DataFrame()
    calculated_summary["Trial 123"] = concat_summary.iloc[:num_var,:3].sum(axis=1)
    calculated_summary["Trial 456"] = concat_summary.iloc[:num_var,3:6].sum(axis=1)
    calculated_summary["Trial 789"] = concat_summary.iloc[:num_var,6:].sum(axis=1)
    ANOVA_sum_table.loc[i] = np.array(calculated_summary).reshape(1,-1)
    # print("hi")
    

ANOVA_mean_table.to_excel(result_addr+"mean"+filename)
ANOVA_median_table.to_excel(result_addr+"median"+filename)
ANOVA_min_table.to_excel(result_addr+"min"+filename)
ANOVA_max_table.to_excel(result_addr+"max"+filename)
ANOVA_sum_table.to_excel(result_addr+"sum"+filename)
ANOVA_std_table.to_excel(result_addr+"std"+filename)
ANOVA_cv_table.to_excel(result_addr+"cv"+filename)
```

```python
#전체 trial 뽑는 코드

ANOVA_concat_table= pd.DataFrame(columns=column_names, index= sheet_names)

for i in sheet_names:
    summary = df[i][summary_start_row +2*num_var+1 +1:summary_start_row+3*num_var +1 +1]
    summary = summary.replace("#DIV/0!", pd.NA)
    summary.iloc[:5,range(summary_start_column, summary.shape[1], summary_trial_offset)] /= 1000 
    concat_summary = pd.DataFrame()
    for j in range(summary_start_column, summary.shape[1], summary_trial_offset):
        concat_summary = pd.concat([concat_summary, summary.iloc[:,j]], axis=1)
    ANOVA_concat_table.loc[i] = np.array(concat_summary).reshape(1,-1)

ANOVA_concat_table.to_excel(result_addr+"entire"+filename)
```
