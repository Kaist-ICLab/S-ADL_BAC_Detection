```python
import pandas as pd
import numpy as np
```

```python
filename = "Univariate.xlsx"
file_addr = "./HCI_Data/ANOVA_SPSS_sum/"
result_addr = "./HCI_Data/result/"
```

```python
df = pd.read_excel(file_addr+filename, engine='openpyxl')


row_names =[
"total_log",
"Input_Stream__(C+IF+INF+F)",
"Fix__(F)",
"Incorrect_Fixed__(IF)",
"Incorrect_Not_Fixed__(INF)",
"All_Actions__(shift, 한영, 특수문자)",
"Input_Stream_with_all_actions__(ISø)",
"total_time__(Seconds)",
"Transcripted_text__(T)",
"Word_per_second__(WPS)",
"Word_per_minute__WPM)",
"AdjWPS",
"AdjWPM",
"Character_per_second__(CPS)",
"Keystrokes_per_second__(KSPS)",
"Gesture_per_second_(GPS)",
"average_intercharacter_time",
"median_intercharacter_time",
"min_intercharacter_time",
"max_intercharacter_time",
"Keystrokes_per_character__(KSPC)",
"Gesture_per_character__(GPC)",
"Minimum_string_distance__(MSD)",
"Corrected_error_rates",
"Uncorrected_error_rates",
"Total_error_rates",
"Utilized_bandwidth",
"Wasted_bandwidth",
"Cost_per_correction",
]
column_names =["Mauchly’s Test of Sphericity (P>0.05)","Test of Within-Subjects Effects (P<0.05)","Multivariate Tests (P<0.05)","Pairwise Comparisons(0<->0.03)(P<0.05)","Pairwise Comparisons(0<->0.08)(P<0.05)","Pairwise Comparisons(0.03<->0.08)(P<0.05)","Within-Subjects Contrasts(Level 1 vs Level 2)","Within-Subjects Contrasts(Level 2 vs Level 3)"]
```

```python

Uni_ANOVA_table = pd.DataFrame(index= row_names)


#Mauchly
con_Mauchly = df[df.iloc[:,2] == "근사 카이제곱검정"]
new_index=[]
for i in con_Mauchly.index:
    new_index.append(i+2)
Mauchly_table = pd.DataFrame(index= new_index)
for i in new_index:
    Mauchly_table.loc[i,0] = df.iloc[i,4]
Mauchly_table = Mauchly_table.reset_index().drop("index", axis=1)
# print(Mauchly_table)

#With-in
con_within = df[df.iloc[:,0] == "개체-내 효과 검정"]
sphericity_index=[]
for i in con_within.index:
    sphericity_index.append(i+3)
# print(sphericity_index)
within_table = pd.DataFrame(index= sphericity_index)
idx=0
for i in sphericity_index:
    # print(df.iloc[i,6])
    if (Mauchly_table[0][idx]>=0.05):
        within_table.loc[i,0] = df.iloc[i,6]
    else:
        within_table.loc[i,0] = df.iloc[i+1,6]
    idx+=1
within_table = within_table.reset_index().drop("index", axis=1)
# print(within_table)

#Multi variate
con_multi = df[df.iloc[:,0] == "다변량 검정a"]
multi_index=[]
for i in con_multi.index:
    multi_index.append(i+2)
multi_table = pd.DataFrame(index= multi_index)
for i in multi_index:
    multi_table.loc[i,0] = df.iloc[i,6]
multi_table = multi_table.reset_index().drop("index", axis=1)
# print(multi_table)

#pairwise
con_pair = df[df.iloc[:,0] == "대응별 비교"]
pair_index=[]
for i in con_pair.index:
    pair_index.append(i+4)
pair_table = pd.DataFrame(index= pair_index)
for i in pair_index:
    pair_table.loc[i,0] = df.iloc[i,4]
    pair_table.loc[i,1] = df.iloc[i+1,4]
    pair_table.loc[i,2] = df.iloc[i+3,4]
pair_table = pair_table.reset_index().drop("index", axis=1)
# print(pair_table)

#within
con_within_contrast = df[df.iloc[:,0] == "개체-내 대비 검정"]
within_contrast_index=[]
for i in con_within_contrast.index:
    within_contrast_index.append(i+3)
within_contrast_table = pd.DataFrame(index= within_contrast_index)
for i in within_contrast_index:
    within_contrast_table.loc[i,0] = df.iloc[i,6]
    within_contrast_table.loc[i,1] = df.iloc[i+1,6]
within_contrast_table = within_contrast_table.reset_index().drop("index", axis=1)


concat_df= pd.concat([Mauchly_table,within_table,multi_table ,pair_table,within_contrast_table], axis=1, ignore_index=True)
concat_df.index = row_names
concat_df.columns = column_names
print(concat_df)
concat_df.to_excel(result_addr+"sum" +filename)
    
```
