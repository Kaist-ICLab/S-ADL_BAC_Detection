```python

con_friedman = df[df.iloc[:,0] == "검정 통계량a"]

friedman_index=[]
for i in con_friedman.index:
    friedman_index.append(i+3)
# print(friedman_index)
friedman_table = pd.DataFrame(index= friedman_index)
for i in friedman_index:
    friedman_table.loc[i,0]= df.iloc[i,1]
    friedman_table.loc[i,1]= df.iloc[i,2]
    friedman_table.loc[i,2]= df.iloc[i,3]
friedman_table = friedman_table.reset_index().drop("index", axis=1)
print(friedman_table)



friedman_table.to_excel(result_addr+"median"+filename)
    
```
